```python
from pydantic import BaseModel
```

```python
class User(BaseModel):
    id: int
    name: str
    email: str
    is_active: bool = True
```

```python
user = User(id=1, name="Alice", email="alice@example.com", is_active=True)
print(user)
```

```python
user2 = User(id=2, name="Bob", email="bob@example.com", is_active=False)
print(user2)
```

```python
class car(BaseModel):
    id: int
    name: str
    model: str
    models_year: int
    variance: list
    
# Creating an instance of the car class
car1 = car(id=1, name="BMW", model="X5", models_year=2020, variance=["X5 M", "X5 xDrive40i"])
print(car1)
```

```python
# Creating another instance of the car class with input as tuple but converting it to list 
car2 = car(id=2, name="Audi", model="Q7", models_year=2021, variance=("Q7 55 TFSI", "Q7 45 TDI"))
print(car2)
```

```python
class voice_agent(BaseModel):
    id: int
    name: str
    language: str

agent1 = voice_agent(id=1, name="Siri", language="English")
print(agent1)
```

```python
class room(BaseModel):
    id: int
    name: str
    capacity: dict
    
room1 = room(id=1, name="Conference Room", capacity={"max_people": 50, "min_people": 5})
print(room1)
```

```python
class laptop(BaseModel):
    id: int
    brand: str
    specs: dict 
    
laptop1 = laptop(id=1, brand="Dell", specs={"RAM": "16GB", "Storage": "512GB SSD"})
print(laptop1)
```
